```python
# imports 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
# read in data
bos_crime_2023 = pd.read_csv('data/Bos_crime_2023.csv')
bos_crime_2023.head()
```

```python
# create a new df
crime_count_df = bos_crime_2023
# remove time from the date column
crime_count_df['OCCURRED_ON_DATE'] = pd.to_datetime(crime_count_df['OCCURRED_ON_DATE']).dt.date

# count how many unique days there are
print(len(crime_count_df['OCCURRED_ON_DATE'].unique()))

# use total rows to devide by unique days
print(len(crime_count_df)/len(crime_count_df['OCCURRED_ON_DATE'].unique()))

# create a new df remove all rows with 0 in SHOOTING column
no_gun_df = crime_count_df[crime_count_df['SHOOTING'] != 0]

# no_gun_df.head()

print(len(no_gun_df)/len(crime_count_df['OCCURRED_ON_DATE'].unique()))

```

```python
# round to no decimal places
print(round(len(crime_count_df)/len(crime_count_df['OCCURRED_ON_DATE'].unique())))
print(round(len(no_gun_df)/len(crime_count_df['OCCURRED_ON_DATE'].unique())))
```
